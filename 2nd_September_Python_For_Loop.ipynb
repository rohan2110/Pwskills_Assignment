### 20. Write a program to find the greatest common divisor (GCD) of two numbers using a `for` loop.

```python
# Function to find the GCD of two numbers using a for loop
def find_gcd(num1, num2):
    # Ensure num1 is greater than or equal to num2
    if num1 < num2:
        num1, num2 = num2, num1

    # Initialize GCD to 1 (minimum possible value)
    gcd = 1

    # Iterate from 2 to the smaller of the two numbers
    for i in range(2, num2 + 1):
        # Check if i divides both num1 and num2
        if num1 % i == 0 and num2 % i == 0:
            gcd = i  # Update the GCD if i is a common divisor

    return gcd

# Input two numbers from the user
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Calculate and display the GCD
result = find_gcd(num1, num2)
print(f"The GCD of {num1} and {num2} is {result}")
```

### 34. Write a program that calculates the mean, median, and mode of a list of numbers using list comprehension.

```python
l =[10,20,30,40,50,60,70,80,90,90,100]


from collections import Counter
# define a function to calculate mean, median and mode
def mean_median_mode(x):
    x = list(x)
    mean = sum(l)/len(l)
    
    x.sort()
    if (len(x)%2 ==0):
        median1= x[(len(x)//2)]
        median2= x[(len(x)//2)-1]
        median = (median1 + median2)/2
    else:
        median = x[(len(x))//2]
        
    counter = Counter(x)
    max_count = max(counter.values())
    mode = [num for num, count in counter.items() if count == max_count]
    
    return [mean,median,mode]

print(f"The Mean,Median and Mode for list '{l}' are ",mean_median_mode(l)," respectively")
```

### 35. Create a program that generates Pascal's triangle up to a specified number of rows using list comprehension.

```python
# Print Pascal's Triangle in Python
 
# input n
n = 5
 
for i in range(1, n+1):
    for j in range(0, n-i+1):
        print(' ', end='')
 
    # first element is always 1
    C = 1
    for j in range(1, i+1):
 
        # first value in a line is always 1
        print(' ', C, sep='', end='')
 
        # using Binomial Coefficient
        C = C * (i - j) // j
    print()
```
